```
#Resolver a vírgula como separador de decimal
preco_remedio['PF Sem Impostos'] = preco_remedio['PF Sem Impostos'].apply(lambda x: str(x).replace(',','.'))
```

```
preco_remedio['PF Sem Impostos']
```

```
#Convetendo a coluna para float
preco_remedio['PF Sem Impostos'] = preco_remedio['PF Sem Impostos'].astype('float64')
```
