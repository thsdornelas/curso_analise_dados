```python
import pandas as pd
import matplotlib as plt
```

```python
colunas_selecionadas = ['NU_INSCRICAO','NU_NOTA_MT', 'NU_NOTA_REDACAO', 'Q001', 'Q002']
```

```python
microdadosEnem2019 = pd.read_csv('D:/temp/programacao_dinamica/MICRODADOS_ENEM_2019.csv', sep=';', encoding='ISO-8859-1',
                                usecols = colunas_selecionadas)
```

```python
microdadosEnem2019.head()
```

```python
microdadosEnem2019 = microdadosEnem2019.dropna()
```

```python
microdadosEnem2019.head()
```

```python
#Introdução ao conceito de dicionário
q001e002Dicionario = {'A': 'Nunca estudou',
                      'B': 'Não completou a 4ª série/5º ano do ensino fundamental',
                      'C': 'Completou a 4ª séria/5° ano do ensino fundamental',
                      'D': 'Completou a 8ª serie/9° ano do ensino fundamental',
                      'E': 'Completou o ensino médio, mas não completou a faculdade',
                      'F': 'Completou a faculdade, mas não completou a pós-graduação',
                      'G': 'Completou a pós-graduação',
                      'H': 'Não sei'
}
```

```python
q001e002Dicionario['A']
```

```python
microdadosEnem2019.Q001
```

```python
#Inserir uma nova coluna
microdadosEnem2019['NO_Q001'] = [q001e002Dicionario[resp] for resp in microdadosEnem2019.Q001]
```

```python
microdadosEnem2019.head()
```

```python
microdadosEnem2019['NO_Q002'] = [q001e002Dicionario[resp] for resp in microdadosEnem2019.Q002]
```

```python
microdadosEnem2019.head()
```

```python
microdadosEnem2019.filter(items = ['NU_INSCRICAO', 'NO_Q001']).groupby('NO_Q001').count().sort_values(by='NU_INSCRICAO', ascending=False)
```

```python
microdadosEnem2019.filter(items = ['NU_INSCRICAO', 'NO_Q002']).groupby('NO_Q002').count().sort_values(by='NU_INSCRICAO', ascending=False)
```

```python
microdadosEnem2019.filter(items = ['NU_NOTA_MT', 'NO_Q001']).groupby('NO_Q001').mean().sort_values(by='NU_NOTA_MT', ascending=False)
```

```python
microdadosEnem2019.filter(items = ['NU_NOTA_MT', 'NO_Q002']).groupby('NO_Q002').mean().sort_values(by='NU_NOTA_MT', ascending=False)
```

```python
microdadosEnem2019.filter(items = ['NU_NOTA_REDACAO', 'NO_Q001']).groupby('NO_Q001').mean().sort_values(by='NU_NOTA_REDACAO', ascending=False)
```

```python
microdadosEnem2019.filter(items = ['NU_NOTA_REDACAO', 'NO_Q002']).groupby('NO_Q002').mean().sort_values(by='NU_NOTA_REDACAO', ascending=False)
```
